**We'll be using Random Forest for predicting the housing prices!**

**Firstly, we'll import the data set using pandas**

```python
train_data = pd.read_csv('../input/home-data-for-ml-course/train.csv')
test_data = pd.read_csv('../input/home-data-for-ml-course/test.csv')
```

**Displaying the first few features to decide our features and target**

```python
train_data.head()
```

```python
test_data.head()
```

**Choosing our features and target for predictions**

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train_data[features]
X.head()
```

```python
y = train_data.SalePrice
```

**Splitting the training data to build the model using Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=1)
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_model_prediction = rf_model.predict(test_X)
rf_mae = mean_absolute_error(rf_model_prediction, test_y)
print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_mae))
```

**Now, building a model for full data prediction**

```python
rf_full = RandomForestRegressor()
rf_full.fit(X, y)
testX = test_data[features]
test_preds = rf_full.predict(testX)
```

**Converting to CSV file**

```python
output = pd.DataFrame({'Id': test_data.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
